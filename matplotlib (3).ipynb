```
x1=np.linspace(-np.pi,np.pi,256)
x2=np.linspace(-np.pi,np.pi,256)

y1=np.cos(x1)
y2=np.sin(x2)

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1 ,'r--')
plt.plot(x2,y2 ,'b-.')
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue')
plt.plot(x2,y2,color='brown')
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue',lw=3.6)
plt.plot(x2,y2,color='brown',lw=3.6)
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue',lw=3.6,ls='--')
plt.plot(x2,y2,color='brown',lw=3.6,ls='-.')
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue',lw=3.6,ls='--')
plt.plot(x2,y2,color='brown',lw=3.6,ls='-.')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])

plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue',lw=3.6,ls='--')
plt.plot(x2,y2,color='brown',lw=3.6,ls='-.')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])
plt.show()
```

```
plt.figure(figsize=(8,5),dpi=100)
plt.plot(x1,y1,color='blue',lw=3.6,ls='--',label='$y=cosx$')
plt.plot(x2,y2,color='brown',lw=3.6,ls='-.',label='$y=sin$')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1,0,1])

plt.legend(loc=0)
plt.show()
```

```
x=np.array([1,2,3,4,5])
y1=x**2
y2=np.exp(x)
plt.figure(figsize=(8,5),dpi=100)
plt.subplot(2,1,1)
plt.plot(x,y1,'p')

plt.subplot(2,1,2)
plt.plot(x,y2,'y--')
```

```
x=np.array([2,4,6,8,10,12,14,15,16,18,20])
y=x**4
fig=plt.figure(figsize=(8,5))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```
fig=plt.figure(figsize=(10,6),dpi=100)
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.15,0.55 ,0.4,0.3])
axes1.plot(x,y, 'r--')
axes1.set_title('large plot')
axes2.plot(y,x, 'g-.')
axes2.set_title('small plot')
plt.show()
```

```
t= np.arange(0,3,0.01)
s=1+np.sin(2*np.pi*t)
fig,axes=plt.subplots()
axes.plot(t,s)
axes.set(xlabel='time',ylabel='voltages')
axes.grid()
plt.show()
```
