```
df = pd.read_csv(r'/content/car price prediction.csv')
df.head()
```

```
df['num-of-doors'] = df['num-of-doors'].fillna('Zero')
```

```
df
```

```
df['num-of-doors'] = df['num-of-doors'].apply(w2n.word_to_num)
df
```

```
import math
median_test_score = math.floor(df['bore'].mean())
median_test_score
```

```
dff = df['bore'].mean()
dff
```

```
df['bore'] = df['bore'].fillna(dff)
df
```

```
df.columns
```

```
predictors = ['num-of-doors', 'city-mpg', 'peak-rpm']
x = df[predictors]
y = df['highway-mpg']
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```
x_train.shape,x_test.shape
```

```
y_train.shape,y_test.shape
```

```
reg = LinearRegression()
model = reg.fit(x_train,y_train)
model.predict([[8,6,7]])
```

```
model.score(x_train,y_train)
```

```
model.score(x_test,y_test)
```
