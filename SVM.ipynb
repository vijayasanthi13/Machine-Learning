```
x = df.drop(['Class'],axis=1)
x
```

```
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```
svmAlgo = SVC()
```

```
model = svmAlgo.fit(x_train, y_train)
```

```
model.score(x_train, y_train)
```
