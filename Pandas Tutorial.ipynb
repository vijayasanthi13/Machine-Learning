# Merging, Joining and Concatenating

## Concatenating

```python
df1 = pd.DataFrame(np.random.randint(1, 20, size = (4, 3)), [1,2,3,4], ['W', 'X', 'Y'])
```

```python
df2 = pd.DataFrame(np.random.randint(20,40, size = (4, 3)), [5,6,7,8], ['W', 'X', 'Y'])
```

```python
df3 = pd.DataFrame(np.random.randint(40,60, size = (4, 3)), [9,10,11,12], ['W', 'X', 'Y'])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
farme = [df1, df2, df3]
```

```python
pd.concat(farme)
```

```python
pd.concat(farme, axis = 1)
```

## Merging

```python
left = {
    'key':['k0','k1','k2','k3'],
    'A':[10,20,30,40],
    'B':[1,2,3,4]
}
```

```python
right = {
    'key':['k0','k1','k2','k3'],
    'C':[100,200,300,400],
    'D':[11,12,13,14]
}
```

```python
left = pd.DataFrame(left)
```

```python
right = pd.DataFrame(right)
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge
```

## Joining

```python
left = {
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
}
```

```python
right = {
    'C':['C0','C2','C3'],
    'D':['D0','D2','D3']
}
```

```python
left = pd.DataFrame(left, index = ['k0','k1','k2'])
```

```python
right = pd.DataFrame(right, index = ['k0','k2','k3'])
```

```python
left
```

```python
right
```

```python
left.join(right, how = 'inner')
```

```python
left.join
```
